## Data Preprocessing

#### Standardization

```python
sc = StandardScaler()

# Replaces unstandardized columns with standardized columns
for col in X.columns:
    # Train data
    temp = np.array(X[col]).reshape(-1, 1)
    X[col] = sc.fit_transform(temp)
    
    # Test data
    temp2 = np.array(test_data[col]).reshape(-1, 1)
    test_data[col] = sc.fit_transform(temp2)
    
# Splits data into training and validation sets
X_train = X.iloc[:700,:]
y_train = y.iloc[:700,:]
X_val = X.iloc[700:,:]
y_val = y.iloc[700:,:]
```

## Modeling

#### Creates Precision, Recall & Accuracy Functions to Evaluate Performance

```python
def precision(y, y_preds):
    '''Returns precision, which is TP/(TP+FP).'''
    
    tp, fp = 0,  0
    for i in range(len(y_preds)):
        if y_preds[i] == 1:
            if y[i] == 1:
                tp += 1
            else:
                fp += 1
    
    return tp / (tp+fp) if tp+fp > 0 else 0
```

```python
def recall(y, y_preds):
    '''Returns recall, which is TP/(TP+FN).'''

    tp, fn = 0,  0
    for i in range(len(y_preds)):
        if y_preds[i] == 1 and y[i] == 1:
            tp += 1
        elif y[i] == 1 and y_preds[i] == 0:
            fn += 1
                
    return tp / (tp+fn)
```

```python
def accuracy(y, y_preds):
    '''Returns accuracy, which is (TP+TN)/(TP+FP+FN+TN).'''

    tp, tn, fp, fn = 0, 0, 0, 0
    
    for i in range(len(y_preds)):
        if y_preds[i] == 1 and y[i] == 1:
            tp += 1
        elif y_preds[i] == 1 and y[i] == 0:
            fp += 1
        elif y_preds[i] == 0 and y[i] == 1:
            fn += 1
        elif y_preds[i] == 0 and y[i] == 0:
            tn += 1
                
    return (tp + tn) / (tp + fp + fn + tn)
```

#### Decision Tree

```python
'''Builds Decision Tree Model.'''

decision_tree = DecisionTreeClassifier(random_state=20)
decision_tree = decision_tree.fit(X_train, y_train)

# Makes predictions on validation data
y_val_pred = decision_tree.predict(X_val)
y_val = np.array([i[0] for i in y_val.values.tolist()])

# Prints performance
print(f'Accuracy:  {accuracy(y_val, y_val_pred):.5f}')
print(f'Precision: {precision(y_val, y_val_pred):.5f}')
print(f'Recall:    {recall(y_val, y_val_pred):.5f}')
```

#### Random Forest

```python
'''Builds Random Forest Model.'''

random_forest = RandomForestClassifier(random_state=20)
random_forest.fit(X_train, y_train)

# Makes predictions on validation data
rf_preds = random_forest.predict(X_val)

# Prints performance
print(f'Accuracy:  {accuracy(y_val, rf_preds):.5f}')
print(f'Precision: {precision(y_val, rf_preds):.5f}')
print(f'Recall:    {recall(y_val, rf_preds):.5f}')
```

```python
'''Plots ROC Curve.'''

rf_probs = random_forest.predict_proba(X_val)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_val, rf_probs, pos_label=1)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1]), plt.ylim([0, 1])
plt.ylabel('True Positive Rate'), plt.xlabel('False Positive Rate')
plt.show()
```

**The Random Forest model is better than the Decision Tree model. This is indicated by the greater accuracy, precision and recall value for the Random Forest model. Furthermore, the AUC value for the Random Forest model (0.79) is greater than the Decision Tree AUC value (0.61).**

## Tuning Hyperparameters

```python
'''Performs Grid Search.'''

param_grid = {
    'max_depth': [None],
    'max_features': ['log2'],
    'min_samples_leaf': [1, 2, 3, 4, 5],
    'min_samples_split': [2, 3, 4, 5],
    'n_estimators': [5, 12, 15, 18, 22]
}

random_forest_hyp = RandomForestClassifier(random_state=20)

grid_search = GridSearchCV(estimator=random_forest_hyp, scoring='f1', param_grid=param_grid, 
                           cv=3, n_jobs=-1, verbose=2)
grid_search = grid_search.fit(X_train, y_train.quality.ravel())
```

```python
print(f'The Best Parameters are:\n')
[print(f'{k:<18}-->\t{v}') for k, v in grid_search.best_params_.items()];
```

```python
print(f'The Best Score Is: {grid_search.best_score_:.4f}')
```

```python
'''Predicts on Validation Set Using Optimal Hyperparameters.'''

random_forest_best = RandomForestClassifier(random_state=20, max_depth=None, max_features='log2',
                                            min_samples_leaf=4, min_samples_split=2, n_estimators=22)

random_forest_best = random_forest_best.fit(X_train, y_train)
random_forest_preds = random_forest_best.predict(X_val)

# Predicts on validation set
print(f'Accuracy:  {accuracy(y_val, random_forest_preds):.5f}')
print(f'Precision: {precision(y_val, random_forest_preds):.5f}')
print(f'Recall:    {recall(y_val, random_forest_preds):.5f}')
```

```python
'''Plots ROC Curve.'''

rf_best_probs = random_forest_best.predict_proba(X_val)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_val, rf_best_probs, pos_label=1)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1]), plt.ylim([0, 1])
plt.ylabel('True Positive Rate'), plt.xlabel('False Positive Rate')
plt.show()
```
